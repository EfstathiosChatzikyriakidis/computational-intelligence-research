# A simple program implementing a basic menu of operations

```
print("Welcome to the nature center. What would you like to do?")

choice = ''

while choice != 'q':
    print("[1] Enter 1 to take a bicycle ride.")
    print("[2] Enter 2 to go for a run.")
    print("[3] Enter 3 to climb a mountain.")
    print("[q] Enter q to quit.")
    
    choice = input("What would you like to do? ")
    
    if choice == '1':
        print("Here's a bicycle. Have fun!")
    elif choice == '2':
        print("Here are some running shoes. Run fast!")
    elif choice == '3':
        print("Here's a map. Can you leave a trip plan for us?")
    elif choice == 'q':
        print("Thanks for playing. See you later.")
    else:
        print("I don't understand that choice, please try again.")

print("Thanks again, bye now.")
```

# A simple program for finding the right word!

```
python_words = {
    'list'       : 'A collection of values that are not connected, but have an order.',
    'dictionary' : 'A collection of key-value pairs.',
    'function'   : 'A named set of instructions that defines a set of actions in Python.'
}

for meaning in python_words.values():
    print("Meaning: %s" % meaning)
    
    guessed_word = input("What word do you think this is? ")
    
    if guessed_word in python_words and python_words[guessed_word] == meaning:
        print("You got it!")
    else:
        print("Sorry, that's just not the right word.")
```

```
def get_number_word(number):
    if number == 0:
        return 'zero'
    elif number == 1:
        return 'one'
    elif number == 2:
        return 'two'
    elif number == 3:
        return 'three'
    else:
        return "I'm sorry, I don't know that number."

test = get_number_word(3)

print(test)

for current_number in range(0, 5):
    print(current_number, get_number_word(current_number))
```

```
def double(x):
    return 2*x

def triple(x):
    return 3*x

def double_and_triple(x):
    return (2*x, 3*x)

result = double(3)
print(result)

result = triple(3)
print(result)

result = double_and_triple(3)
print(result)

a, b = double_and_triple(3)
print(a)
print(b)
```

# A beatiful way to call function arguments

```
def describe_person(first_name, last_name, age):
    print("First name: %s" % first_name.title())
    print("Last name: %s" % last_name.title())
    print("Age: %d\n" % age)

describe_person(age=71, first_name='brian', last_name='kernighan')
describe_person(age=70, first_name='ken', last_name='thompson')
describe_person(age=68, first_name='adele', last_name='goldberg')
```

# Use optional arguments in a function

```
def describe_person(first_name, last_name, age=None, favorite_language=None):
    print("First name: %s" % first_name.title())
    print("Last name: %s" % last_name.title())

    if age: print("Age: %d" % age)

    if favorite_language: print("Favorite language: %s" % favorite_language)

    print()

describe_person('adele', 'goldberg', favorite_language='C')
describe_person('adele', 'goldberg', age=68)
describe_person('adele', 'goldberg', age=68, favorite_language='Smalltalk')
describe_person('adele', 'goldberg')
```

#Accept unlimited number of arguments in a function

```
def example_function(arg_1, arg_2, *more_arguments):
    print('arg_1:', arg_1)
    print('arg_2:', arg_2)

    for value in more_arguments:
        print('one more value:', value)

    print()

example_function(1, 2)
example_function(1, 2, 3)
example_function(1, 2, 3, 4)
example_function(1, 2, 3, 4, 5)
```

```
def adder(*nums):
    s = 0 

    for num in nums:
        s = s + num

    print("The sum of your numbers is %d." % s)
    
adder(1, 2, 3)

print(sum([1, 2, 3]))
```

# Function that accepts 2 positional arguments and infinite (key, value) arguments

```
def example_function(arg_1, arg_2, **kwargs):
    print('arg_1:', arg_1)
    print('arg_2:', arg_2)
    
    for key, value in kwargs.items():
        print('one more (key, value) pair: (' + key + ', ' + value + ')')

    print()

example_function('a', 'b')
example_function('a', 'b', value_3='c')
example_function('a', 'b', value_3='c', value_4='d')
example_function('a', 'b', value_3='c', value_4='d', value_5='e')
```

# Usage of 'filter' with custom function or lambda expresssion to filter out values

```
values = ['A-1: Test', 'A-2: Test', 'B-1: Test']

def filter_values(value):
    return value.startswith('A')

filtered_values = filter(filter_values, values)

for value in filtered_values:
    print(value)

print()

filtered_values = filter(lambda a_str: a_str.startswith('A'), values)

for value in filtered_values:
    print(value)
```

# Usage of 'map' with custom function or lambda expression to transform values

```
values = [1, 2, 3, 4, 5]

def transform_function(value):
    return value * 100

transformed_values = map(transform_function, values)

for value in transformed_values:
    print(value)

print()

transformed_values = map(lambda o: o * 100, values)

for value in transformed_values:
    print(value)
```
